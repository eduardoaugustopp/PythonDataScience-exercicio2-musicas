Importando a biblioteca [`pandas`](https://pandas.pydata.org/docs/index.html) e os dados em um arquivo excel que serão utilizados.

```python
import pandas as pd
```

```python
musicas_albuns = pd.read_excel('/content/drive/MyDrive/pandas-exercicio2-musica.xlsx', sheet_name = 'Popularidade de Músicas')
```

Checando a estrutura do DataFrame com o uso do método [`info()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.info.html?highlight=info#pandas.DataFrame.info), que retorna a quantidade de linhas e colunas, bem como a informação de quantidade de dados nulos e tipos de dados para cada uma das colunas.

```python
musicas_albuns
```

```python
musicas_albuns.info()
```

Verificando os dados dos álbuns que não pertencem à banda Nirvana.

```python
(musicas_albuns['Nirvana / 2pac'] == 'Greatest Hits') | (musicas_albuns['Nirvana / 2pac'] == 'All Eyez on Me')
```

Identificando se o álbum Bleach do Nirvana tem certificação.

```python
musicas_albuns.loc[musicas_albuns['Nirvana / 2pac'] == 'Bleach', 'Certificações'].unique()
```

```python
musicas_albuns = musicas_albuns[musicas_albuns['Nirvana / 2pac'] == 'Nevermind']
musicas_albuns
```

Deletando a coluna Gênero.

```python
musicas_albuns = musicas_albuns.drop(columns= 'Gênero')
musicas_albuns
```

Organizando Ano e Reproduções, modificando o formato da base de dados de forma que as colunas referentes a cada ano se tornem uma só coluna.

```python
#selecionar as linhas que não tem alteração
musicas_albuns.loc[:, 'Música':'Nível de popularidade'].columns
```

```python
colunas_info = list(musicas_albuns.loc[:, 'Música':'Nível de popularidade'].columns)
colunas_info
```

```python
musicas_albuns.loc[:,'Reproduções (milhões) 2005':'Reproduções (milhões) 2010'].columns
```

```python
colunas_musica = list(musicas_albuns.loc[:,'Reproduções (milhões) 2005':'Reproduções (milhões) 2010'].columns)
colunas_musica

```

Realizando a transformação do DataFrame, utilizando o método [`melt()`](https://pandas.pydata.org/docs/reference/api/pandas.melt.html). Esse método tem o intuito de transformar um DataFrame de um formato amplo (*wide*) para o formato longo (*long*).

```python
musicas_albuns.melt(id_vars = colunas_info, value_vars = colunas_musica, var_name = 'Ano' , value_name = 'Reproduções')
```

```python
musicas_por_ano = musicas_albuns.melt(id_vars = colunas_info, value_vars = colunas_musica, var_name = 'Ano' , value_name = 'Reproduções')
```

Análisando a quantidade de reproduções da música durante os anos.


```python
musicas_por_ano.groupby('Música')
```

Checando quais dados foram agrupados em cada grupo com uso do atributo [`groups`](https://pandas.pydata.org/docs/reference/api/pandas.core.groupby.GroupBy.groups.html) e realizar seleções de grupos específicos com o método [`get_group()`](https://pandas.pydata.org/docs/reference/api/pandas.core.groupby.GroupBy.get_group.html).

```python
musicas_por_ano.groupby('Música').groups
```

```python
musicas_por_ano.groupby('Música').get_group('Smells Like Teen Spirit')
```
